```python
def loadCSV(input_file):
    data = []
    
    textfile = open(input_file)
    for line in textfile:
        row_data = line.strip("\n").split()
        for i, item in enumerate(row_data):
                row_data[i] = float(item)
        data.append(row_data)
       
    return np.asarray(data)
```
